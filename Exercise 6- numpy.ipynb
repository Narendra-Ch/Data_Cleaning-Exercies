```python
import numpy  as np
```

```python
 
def vander_monde(in_vect,n,increase=False):
    
    if not increase:
        result=np.array([j**(n-i-1) for j in in_vect for i in range(n)]).reshape(in_vect.size,n)
    elif increase:
        result=np.array([j**i for j in in_vect for i in range(n)]).reshape(in_vect.size,n)
    
    return result

x=np.array([1,2,3,4])
print(vander_monde(x,3,False))
print('\n')
x=np.array([2,3,4,5,6])
print(vander_monde(x,4,True))
                
```

```python
x=np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])
k=3
```

```python
#no.of values in moving average are; n-k+1
n=(x.size)-k+1
n
```

```python
result=[]
for i in range(n):
    result.append((x[i]+x[i+1]+x[i+2])/k)
result=[round(num,2) for num in result]
result
```
